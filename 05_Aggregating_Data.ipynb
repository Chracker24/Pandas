### Helps bring out interesting correlations

```python
pivot = jee.pivot_table(columns=["location_type"], index='family_income', values="dropout", aggfunc='count')
pivot
```

```python
pivot = jee.pivot_table(columns=["location_type"], index='family_income', values="peer_pressure_level",aggfunc=lambda x: x.mode().iloc[0] if not x.mode().empty else None)
pivot
```
